```python
###we will check whether a distribution is guassian or not
##QQplot (quantile quantile plot)
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import seaborn as sns
```

```python
import statsmodels.api as sm 
```

```python
import pylab as py
```

```python
data_points = np.random.normal(0,1,100) #random data points generated
                             #(mean=0,s.d=1)  #standard normal distribution
```

```python
data_points
```

```python
sm.qqplot(data_points,line = '45')
py.show()
```

```python
sns.displot(data_points)
```

```python
sns.displot(data_points,kind='kde')
```

```python
#lognormal distribution
```

```python
gfg=np.random.lognormal(0,1,100)
```

```python
gfg
```

```python
sm.qqplot(gfg,line='45')
py.show()
```

```python
sm.qqplot(np.log(gfg),line='45')  #making staright line
py.show()
```

```python
sns.displot(gfg)
```

```python
sns.displot(gfg,kind='kde')
```
